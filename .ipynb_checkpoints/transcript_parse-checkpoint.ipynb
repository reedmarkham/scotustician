# Libraries, functions

```python
import requests
import pandas as pd
from datetime import datetime
import time

def call_api(url):
    print(f"Getting {url}")
    response = requests.get(url)
    parsed = response.json()
    return parsed
```

# Table of cases

```python
case_summaries_json = call_api('https://api.oyez.org/cases?per_page=0')

cases = []

for case_summary in case_summaries_json:
    df = pd.json_normalize(case_summary)
    cases.append(df)
    
cases_df = pd.concat(cases)
case_summary_df = cases_df[['ID','name','href','docket_number','term']].set_index('ID')
case_summary_df.to_csv('data/case_summary.csv', index=True)
```

# Table of oral arguments

```python
case_summary_df = pd.read_csv('data/case_summary.csv')

print("Case table ready: ",datetime.now())

recent_cases = case_summary_df[(case_summary_df['term'] >= "1970") & (case_summary_df['term'] <= "1994")]

oa_list = []

for index, case in recent_cases.iterrows():
    time.sleep(5)
    case_json = call_api(case['href'])
    if case_json['unavailable'] != True and case_json['oral_argument_audio'] is not None:
        case_id = case_json['ID']
        oa = pd.json_normalize(case_json['oral_argument_audio'])[['id','title','href']].set_index('id')
        oa['case_id'] = case_id
        oa = oa[['case_id','title','href']]
        oa_list.append(oa)

oa_df = pd.concat(oa_list)
oa_df.to_csv('data/oa_summary.csv', index=True)
```

```python
oa_df = pd.concat(oa_list)
oa_df.to_csv('data/oa_summary.csv', index=True)
```

# Table of speech (parsed transcripts)

```python
oa_summary_df = pd.read_csv('data/oa_summary.csv')
oa_summary_df = oa_summary_df[(oa_summary_df['title'] != 'No oral argument') & (oa_summary_df['title'] != 'Oral argument unavailable')]
oa_summary_df = oa_summary_df.sort_values(by=['id'])
oa_summary_df.head()
```

```python
petitioner_transcripts = []

for index, oa in oa_summary_df.iterrows():
    time.sleep(5)
    oa_json = call_api(oa['href'])        
    oa_id = oa_json['id']
    if oa_json['transcript'] is not None:
        transcript = oa_json['transcript']
        petitioner_transcript = transcript['sections'][0]
        for index, turn in enumerate(petitioner_transcript['turns']):
            for speech in turn['text_blocks']:
                start = speech['start']
                stop = speech['stop']
                text = speech['text']
                if turn['speaker'] is not None:
                    speaker_id = turn['speaker']['ID']
                else:
                    if "(Inaudible)" in text:
                        speaker_id = "N/A"
                    else:
                        if petitioner_transcript['turns'][index - 2]['speaker'] is None:
                            speaker_id = "N/A"
                        else:
                            speaker_id = petitioner_transcript['turns'][index - 2]['speaker']['ID']
            pt_dict = {"oa_id":oa_id,
                       "oa_section:","petitioner"
                       "speaker_id":speaker_id,
                       "start":start,
                       "stop":stop,
                       "text":text
                      }
            petitioner_transcripts.append(pt_dict)

petitioner_transcript_df = pd.DataFrame(petitioner_transcripts)
```
